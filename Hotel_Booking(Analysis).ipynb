# **Outlier removing of lead_time**

```python
## Outlier removing of lead_time
Q3 = df['lead_time'].quantile(0.75)
Q1 = df['lead_time'].quantile(0.25)
IQR = Q3 - Q1
Lower_Bound = Q1 - 1.5 * IQR
Upper_Bound = Q3 + 1.5 * IQR
df_no_outliers = df[(df['lead_time'] >= Lower_Bound) & (df['lead_time'] <= Upper_Bound)]
```

#:📌 ***DATA_VISUALIZATIONS***

```python
reservation_status_counts = df['is_canceled'].value_counts()
print(reservation_status_counts)

plt.figure(figsize=(4,3))
plt.title('Reservation_Status_Counts')

plt.bar(['Not_Canceled','Canceled'], reservation_status_counts.values, color=['green','red'])
plt.show()
```

```python
canceled_data_hotel = df.groupby('hotel')['is_canceled'].value_counts(normalize=True).mul(100).rename('percentage').reset_index()
plt.figure(figsize=(6,4))
sns.barplot(x='hotel', y='percentage', hue='is_canceled', data=canceled_data_hotel)
plt.title('Cancellation Percentage by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Cancellation Percentage')
plt.legend(labels=['Not Canceled', 'Canceled'], title='Reservation Status')
plt.show()
```

```python
monthly_adr = df.groupby('arrival_date_month')['adr'].mean().reset_index()

# Define the order of months
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_adr['arrival_date_month'] = pd.Categorical(monthly_adr['arrival_date_month'], categories=month_order, ordered=True)
monthly_adr = monthly_adr.sort_values('arrival_date_month')

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_adr, x='arrival_date_month', y='adr', marker='o')
plt.title('Average Daily Rate (ADR) per Month')
plt.xlabel('Arrival Month')
plt.ylabel('Average Daily Rate (ADR)')

plt.grid(True, linestyle='--')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='arrival_date_month', hue='is_canceled', order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
plt.title('Reservation Status per Month')
plt.xlabel('Arrival Month')
plt.ylabel('Number of Reservations')
plt.legend(labels=['Not Canceled', 'Canceled'], title='Reservation Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
